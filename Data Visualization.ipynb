```python
import matplotlib.pyplot as plt
```

```python
#x axis
x=[1,2,3,4,5]
#y axis
y=[45,46,47,48,49]
```

```python
#plotting  the points
plt.plot(x,y)
```

```python
plt.xlabel("sales")
plt.ylabel("profit")
plt.plot(x,y)
```

```python
plt.title("The Graph")
plt.xlabel("sales")
plt.ylabel("profit")
plt.plot(x,y)
```

```python
#plotiing dotted graph
plt.title("The Graph with dotted ")
plt.xlabel("sales")
plt.ylabel("profit")
plt.plot(x,y,linestyle='dotted')
```

```python
#changing color 
plt.title("The Graph")
plt.xlabel("sales")
plt.ylabel("profit")
plt.plot(x,y,color='cyan')
```

```python
#drawing marker
plt.title("The Graph")
plt.xlabel("sales")
plt.ylabel("profit")
plt.plot(x,y,linestyle='dotted',marker='o')

```

```python
#changing size of marker

plt.title("The Graph")
plt.xlabel("sales")
plt.ylabel("profit")
plt.plot(x,y,linestyle='dotted',marker='o',ms=25)
```

```python
#bar plot
plt.title("The Graph")
plt.xlabel("sales")
plt.ylabel("profit")
plt.bar(x,y)
```

```python
#bar plot with line plot
plt.title("The Graph")
plt.xlabel("sales")
plt.ylabel("profit")
plt.plot(x,y,linestyle='dotted',marker='o',ms=5)
plt.bar(x,y)
```

```python
#scatter plot
plt.scatter(x,y)
```
